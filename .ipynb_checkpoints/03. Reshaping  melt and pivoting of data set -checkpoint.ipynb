```python
# Reshaping and pivoting of data set

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import pandas as pd
```

```python
df = pd.read_csv('NYSE_daily_prices_A.csv', parse_dates=['date'])
df.head(5)
```

```python
df.shape

type(df['date'][0])

df['date'][0].year

#df['year'], df['month'] = df['date'].dt.year, df['date'].dt.month
```

```python
# select row which value is stock_symbol is 'AEA'
#df.head(5)
df1 = df[(df['stock_symbol']=='AEA')]
df2 = df[(df['stock_symbol'] == 'ATV')]
df1.head(2)
df2.head(2)
```

```python
df3 = df[(df['stock_symbol']=='AEA') & (df['stock_symbol']=='ATV')]
df3.head(10)
```

```python
#df4 = df[(df['date'].dt.year == 2010) & (df['date'].dt.month == 1) & (df['date'].dt.day == 4)]
#df4

df5 = df[((df['date']== '2010-01-04')|(df['date']== '2010-01-05')) & (df['stock_symbol'].isin(['AEA','ASP','ATV']))]
df5
```

```python
# pivot table
# Here pass the columns name which you want to make as index,
# pass the column name on which you want to make it as pivote table
# in value pass all column on which you wnat 
df6 = df5.pivot_table(index='date',columns='stock_symbol', values=['stock_volume'])
df6
```

```python
# melt function: basically reshape table
nyse_data = {
    'date':['2010-01-01','2010-01-02','2010-01-03'],
    'AEA':[5.42,13.13,19.81],
    'ASP':[4.42,12.13,21.81],
    'AIT':[4.24,11.89,21.31],
}

df7 = pd.DataFrame(nyse_data)
df7
```

```python
# melt function convert column to row based on perticular variable
df8 = pd.melt(df7, id_vars=['date'], var_name='stock_symbol',value_name='stock_price_close')
df8
```
