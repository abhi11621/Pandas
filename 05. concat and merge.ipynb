# In this notebook we will capture concat/merge dataframe with help of pandas library.

```python
import pandas as pd

nyse_data1 = {
   'stock_symbol':['AEA','ASP','AIT'],
    'stock_price_open':[4.42,12.13,21.81],
    'stock_price_close':[4.24,11.89,21.31]
}
df1 = pd.DataFrame(nyse_data1)
df1
```

```python
nyse_data2 = {
    'stock_symbol':['AAP','ACP','AEE'],
    'stock_price_open':[4.42,12.13,21.81],
    'stock_price_close':[4.24,11.89,21.31],
}
df2 = pd.DataFrame(nyse_data2)
df2
```

```python
# concat two dataframe row wise
df3 = pd.concat([df1,df2], axis='index')
df3
```

```python
# concat two dataframe columns wise
df4 = pd.concat([df1,df2], axis='columns')
df4
```

```python
# concat two dataframe columns wise
# here in key concating data frame based on key
df5 = pd.concat([df1,df2], axis='columns', keys=['df1','df2'])
df5
```

```python
df5['df2']
```

```python
# concat row wise
df6 = pd.concat([df1,df2], keys=['df1','df2'])
df6
```

```python
# fetching data based on row wise concatenation
df6.loc['df1']
```

```python
# concat dataframe using series
date = pd.Series(['2010-01-01','2010-01-03','2010-01-03'], name='date')
date
```

```python
df7 = pd.concat([df1,date], axis=1) # here we are adding extra column to dataframe
df7
```

```python
nyse_data3 = {
   'stock_symbol':['AEA','ASP','AIT'],
    'stock_price_open':[4.42,12.13,21.81],
    'stock_price_close':[4.24,11.89,21.31]
}
df8 = pd.DataFrame(nyse_data3)
df8
```

```python
nyse_data4 = {
    'stock_symbol':['AAP','ACP','AIT'],
    'stock_price_open':[5.42,13.13,22.81],
    'stock_price_close':[6.24,12.89,22.31],
}
df9 = pd.DataFrame(nyse_data4)
df9
```

```python
df10 = pd.merge(df8,df9, on='stock_symbol')
df10
```

```python
# inner join
df11 = pd.merge(df8,df9, on='stock_symbol',how='inner')
df11
```

```python
# outer join
df11 = pd.merge(df8,df9, on='stock_symbol',how='outer')
df11
# like that you can use left, right
```

```python
# outer join
df12 = pd.merge(df8,df9, on='stock_symbol',how='outer',indicator=True)
df12
# Here indicator is by default False/ if you mark as True it tell how earch is merged
```

```python
# outer join
df13 = pd.merge(df8,df9, on='stock_symbol',how='outer',indicator=True, suffixes=('_first','_second'))
df13
# Here suffixes tell which column value come from which dataframe
```
