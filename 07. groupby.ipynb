# Groupby

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import pandas as pd
df = pd.read_csv('NYSE_daily_prices_A.csv',parse_dates=['date'])
df.head(5)
```

```python
# what is max stock value for each stock_symbol
dfg = df.groupby('stock_symbol')
dfg
```

```python
for stock_symbol, data in dfg:
    print(stock_symbol)
    print(data)
```

```python
dfg.get_group('AEA').max()
```
